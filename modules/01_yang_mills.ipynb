```python
# ==============================================================================
# MODULE 01: Yang–Mills Existence and Mass Gap
# Contextual Analysis Instrumentation (3 → 4 → 5)
#
# This module does NOT assert a formal proof.
# It documents a structural correspondence between
# experimental mass scale and φ-recursive density structure.
# ==============================================================================

# --- DEPENDENCIES ---
try:
    import numpy as np
    import matplotlib.pyplot as plt
    import json
    import os
    import math
    print("✅ Libraries Detected: numpy, matplotlib, json")
except ImportError:
    raise ImportError("Required libraries missing. Run: pip install -r ../requirements.txt")

import sys

# --- DATA LOGGER ---
def log_results(module_id, module_name, data_summary, validation_dict):
    os.makedirs('../data', exist_ok=True)
    os.makedirs('../validation', exist_ok=True)

    with open(f"../data/{module_id}_{module_name}_results.txt", "w") as f:
        f.write(data_summary)

    with open(f"../validation/{module_id}_{module_name}_metric.json", "w") as f:
        json.dump(validation_dict, f, indent=4)

# --- CONTEXT IMPORT (XA + CONSTANTS) ---
sys.path.append(os.path.abspath('..'))
try:
    from geometry.xa_constant import XA, PHI, PLANCK_H, LIGHT_C
    print(f"✅ Context Restored: XA ({XA:.2e} Hz/kg), φ ({PHI:.6f})")
except ImportError:
    PHI = (1 + math.sqrt(5)) / 2
    PLANCK_H = 6.62607015e-34
    LIGHT_C = 299_792_458
    XA = (PHI * LIGHT_C**2) / PLANCK_H
    print("⚠️ Using local fallback constants.")

# ==============================================================================
# 3️⃣ LEVEL 3 — LOCAL FRAME (PERTURBATIVE PARADOX)
# ==============================================================================
print("\n" + "="*60)
print("3️⃣ LEVEL 3 — LOCAL FRAME OBSERVATION")
print("="*60)

print("Perturbative prediction: m = 0 MeV/c²")
print("Interpretation: Gauge bosons appear massless in a locally constrained frame.")

# ==============================================================================
# 4️⃣ LEVEL 4 — STRUCTURAL PATTERN (REPRESENTATIONAL DENSITY)
# ==============================================================================
print("\n" + "="*60)
print("4️⃣ LEVEL 4 — STRUCTURAL PATTERN RECOGNITION")
print("="*60)

# Experimental reference (external, not derived here)
m_mev_experimental = 1704.14  # MeV/c² (f₀ glueball benchmark)

# π-closure vs φ-recursive density ratio
pi_closure = np.pi * np.sqrt(3)
phi_density_multiplier = pi_closure / PHI  # ≈ 3.363

# φ-recursive mass scale (structural, not predictive)
m_mev_phi_recursive = m_mev_experimental * phi_density_multiplier

print(f"Experimental reference mass: {m_mev_experimental:.2f} MeV/c²")
print(f"π√3 / φ density multiplier: {phi_density_multiplier:.4f}")
print(f"φ-recursive density mass: {m_mev_phi_recursive:.2f} MeV/c²")

print("\nInterpretation:")
print("• π represents circular closure (local containment)")
print("• φ represents recursive self-crossing (density accumulation)")
print("• The mass gap reflects the density relation between these frames")

# ==============================================================================
# 5️⃣ LEVEL 5 — STABILIZED DESCRIPTION
# ==============================================================================
print("\n" + "="*60)
print("5️⃣ LEVEL 5 — STABILIZED DESCRIPTION")
print("="*60)

summary_text = f"""Module 01 — Yang–Mills Contextual Analysis

Experimental benchmark:
  f₀ glueball mass = {m_mev_experimental:.2f} MeV/c²

Structural density relation:
  π√3 / φ multiplier = {phi_density_multiplier:.4f}

φ-recursive density mass:
  {m_mev_phi_recursive:.2f} MeV/c²

Observation:
The Yang–Mills mass gap corresponds to a density transition
between circular (π-closed) and recursive (φ-crossing) representations.

φ does not generate the mass scale.
It explains the density relationship between observed scales.
"""

validation_metrics = {
    "module": "01_yang_mills",
    "experimental_mass_mev": round(m_mev_experimental, 2),
    "phi_density_multiplier": round(phi_density_multiplier, 4),
    "phi_recursive_mass_mev": round(m_mev_phi_recursive, 2),
    "phi_value": round(PHI, 6),
    "validation_basis": "structural_density_relation_relative_to_experimental_anchor",
    "status": "INTELLIGIBLE"
}

log_results("01", "yang_mills", summary_text, validation_metrics)
print("✅ Results written to /data and /validation")

# ==============================================================================
# VISUALIZATION — φ-RECURSIVE PROJECTION (REPRESENTATIONAL)
# ==============================================================================
n = np.arange(1, 1200)
theta = n / PHI
r = n

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(10, 10))
plt.scatter(x, y, c=n, cmap='magma', s=8, alpha=0.6)
plt.title("φ-Recursive Projection (Representational)", fontsize=14)
plt.axis('equal')
plt.axis('off')

plt.figtext(
    0.5, 0.02,
    "© 2026 XYAKANYAA — Representational Visualization",
    ha="center",
    fontsize=10,
    weight="bold"
)

os.makedirs('../visuals', exist_ok=True)
plt.savefig('../visuals/01_yang_mills.jpg', dpi=300, bbox_inches="tight")
plt.show()

print("\nObservation complete.")
print("Visual exported to ../visuals/01_yang_mills.jpg")
```
