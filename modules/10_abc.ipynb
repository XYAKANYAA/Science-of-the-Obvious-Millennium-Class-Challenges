```python
# ==============================================================================
# MODULE 10: ABC Conjecture — Coherence Ratio Diagnostic
# Contextual Analysis Instrumentation (3 → 4 → 5)
# ==============================================================================
# This module does NOT claim a proof of the ABC conjecture.
# It provides a reproducible diagnostic: radical(abc), ABC quality q,
# and a representational interference visualization.
# ==============================================================================

# --- DEPENDENCIES ---
try:
    import numpy as np
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
    import json
    import os
    import math
    print("✅ Libraries Detected: numpy, matplotlib, json")
except ImportError:
    raise ImportError("Missing libraries. Run: pip install -r ../requirements.txt")

import sys

# --- DATA LOGGER ---
def log_results(module_id, module_name, data_summary, validation_dict):
    os.makedirs('../data', exist_ok=True)
    os.makedirs('../validation', exist_ok=True)

    with open(f"../data/{module_id}_{module_name}_results.txt", "w") as f:
        f.write(data_summary)

    with open(f"../validation/{module_id}_{module_name}_metric.json", "w") as f:
        json.dump(validation_dict, f, indent=4)

# --- CONTEXT IMPORT (XYAKANYAA XA RATIO) ---
sys.path.append(os.path.abspath('..'))
try:
    from geometry.xa_constant import XA, PHI
    print(f"✅ Context Restored: XYAKANYAA XA Ratio ({XA:.2e} Hz/kg), φ ({PHI:.6f})")
except ImportError:
    PHI = (1 + math.sqrt(5)) / 2
    XA = 2.19e50
    print("⚠️ Using local definitions for PHI and XA.")

# ==============================================================================
# Utilities
# ==============================================================================
def radical_of_n(n: int) -> int:
    """Product of distinct prime factors of |n|."""
    n = abs(int(n))
    if n == 0:
        return 0
    if n == 1:
        return 1
    factors = set()
    d = 2
    while d * d <= n:
        if n % d == 0:
            factors.add(d)
            while n % d == 0:
                n //= d
        d += 1
    if n > 1:
        factors.add(n)
    prod = 1
    for p in factors:
        prod *= p
    return prod

def abc_quality(a: int, b: int, c: int) -> float:
    """q = log(c) / log(rad(abc)), defined for c>1 and rad(abc)>1."""
    rad = radical_of_n(a * b * c)
    if c <= 1 or rad <= 1:
        return float("nan")
    return float(np.log(c) / np.log(rad))

# ==============================================================================
# 3️⃣ LEVEL 3 — LOCAL FRAME (ADD VS MULT)
# ==============================================================================
print("\n" + "="*60)
print("3️⃣ LEVEL 3 — LOCAL FRAME OBSERVATION")
print("="*60)

# Example triple (kept from your original)
a, b, c = 1, 8, 9
rad_abc = radical_of_n(a * b * c)
q = abc_quality(a, b, c)

print(f"Triple: {a} + {b} = {c}")
print(f"rad(abc) = {rad_abc}")
print(f"ABC quality q = log(c)/log(rad(abc)) = {q:.6f}")
print("Observation: additive simplicity and multiplicative structure can diverge in scale descriptors.")

# ==============================================================================
# 4️⃣ LEVEL 4 — STRUCTURAL PATTERN RECOGNITION (REPRESENTATIONAL INTERFERENCE)
# ==============================================================================
print("\n" + "="*60)
print("4️⃣ LEVEL 4 — STRUCTURAL PATTERN RECOGNITION")
print("="*60)

t_space = np.linspace(0, 2*np.pi, 1000)

wave_a = np.sin(t_space * a * PHI)
wave_b = np.sin(t_space * b * PHI)
wave_c = np.sin(t_space * c * PHI)

interference = wave_a + wave_b - wave_c

# Bounded diagnostics of interference
int_peak = float(np.max(interference))
int_trough = float(np.min(interference))
int_std = float(np.std(interference))
int_rms = float(np.sqrt(np.mean(interference**2)))

print(f"Interference peak: {int_peak:.6f}")
print(f"Interference trough: {int_trough:.6f}")
print(f"Interference RMS: {int_rms:.6f}")
print(f"Interference STD: {int_std:.6f}")
print("Interpretation: interference provides a representational compression of triple relations.")

# ==============================================================================
# 5️⃣ LEVEL 5 — STABILIZED DESCRIPTION (BOUNDED SCORE + THRESHOLD)
# ==============================================================================
print("\n" + "="*60)
print("5️⃣ LEVEL 5 — STABILIZED DESCRIPTION")
print("="*60)

# Operational coherence score (bounded)
coherence_score = float(int_rms / (1.0 + q))

# Explicit threshold for "high-q triple" (diagnostic only)
q_threshold = 1.4
high_quality_under_threshold = bool(q >= q_threshold)

print(f"Operational coherence score: {coherence_score:.6f}")
print(f"Quality threshold: {q_threshold:.2f}")
print(f"High quality under threshold: {high_quality_under_threshold}")

summary_text = f"""Module 10 — ABC (Coherence Ratio Diagnostic)

Triple: {a} + {b} = {c}
rad(abc): {rad_abc}
ABC quality q: {q:.6f}

Interference diagnostics (representational):
peak: {int_peak:.6f}
trough: {int_trough:.6f}
rms: {int_rms:.6f}
std: {int_std:.6f}

Operational coherence score: {coherence_score:.6f}
Quality threshold: {q_threshold:.2f}
High quality under threshold: {high_quality_under_threshold}

Notes:
- This module does not prove the ABC conjecture.
- It reports standard ABC diagnostics (radical, quality) and a representational interference model.
"""

validation_metrics = {
    "module": "10_abc",
    "triple": [int(a), int(b), int(c)],
    "rad_abc": int(rad_abc),
    "abc_quality_q": float(q),
    "interference_peak": int_peak,
    "interference_rms": int_rms,
    "interference_std": int_std,
    "operational_coherence_score": coherence_score,
    "q_threshold": float(q_threshold),
    "high_quality_under_threshold": high_quality_under_threshold,
    "phi": float(PHI),
    "xyakanyaa_xa_ratio_hz_per_kg": float(XA),
    "validation_basis": "radical_and_quality_plus_bounded_interference_diagnostics",
    "status": "diagnostic_complete"
}

log_results("10", "abc", summary_text, validation_metrics)
print("✅ Results written to /data and /validation")

# ==============================================================================
# VISUALIZATION: TOROIDAL RESONANCE PATTERN (REPRESENTATIONAL)
# ==============================================================================
os.makedirs('../visuals', exist_ok=True)

res = 100
u = np.linspace(0, 2*np.pi, res)
v = np.linspace(0, 2*np.pi, res)
U, V = np.meshgrid(u, v)
R_maj, r_min = 4, 1.5

X_tor = (R_maj + r_min*np.cos(U)) * np.cos(V)
Y_tor = (R_maj + r_min*np.cos(U)) * np.sin(V)
Z_tor = r_min*np.sin(U)

# Field intensity (representational)
field_intensity = np.sin(U * a * PHI) + np.sin(V * b * PHI)

# ✅ FIX: use np.ptp(...) for compatibility
field_norm = (field_intensity - np.min(field_intensity)) / (np.ptp(field_intensity) + 1e-12)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(
    X_tor, Y_tor, Z_tor,
    facecolors=plt.cm.magma(field_norm),
    shade=False,
    alpha=0.85
)

ax.set_title("ABC Toroidal Pattern (Representational Visualization)", fontsize=14)
ax.set_axis_off()

# Layout + footer
plt.tight_layout(rect=[0, 0.06, 1, 1])
fig.text(
    0.5, 0.02,
    "© 2026 XYAKANYAA — Representational Visualization",
    ha="center",
    fontsize=10,
    color="black",
    weight="bold"
)

plt.savefig('../visuals/10_abc.jpg', dpi=300, bbox_inches="tight")
plt.show()

print("\nObservation complete.")
print("Visual exported to ../visuals/10_abc.jpg")
```
